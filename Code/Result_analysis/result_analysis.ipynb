```python
def add_analysis_temperature(df):
    df = df.copy()

    if "writer_temperature" in df.columns:
        df["analysis_temperature"] = pd.to_numeric(
            df["writer_temperature"],
            errors="coerce"
        )

    elif "generation_temperature" in df.columns:
        df["analysis_temperature"] = pd.to_numeric(
            df["generation_temperature"],
            errors="coerce"
        )

    elif "temperature" in df.columns:
        df["analysis_temperature"] = pd.to_numeric(
            df["temperature"],
            errors="coerce"
        )

    else:
        raise KeyError(
            "No temperature column found. Expected writer_temperature, generation_temperature, or temperature."
        )

    return df
```

### Create a relationship dataframe from separate dataframes

```python
task_dimension_rename_maps = {
    "Public-event narrative synthesis": {
        "core_event_fidelity": "pns_core_event_fidelity",
        "required_entity_coverage": "pns_required_entity_coverage",
        "key_information_coverage": "pns_key_information_coverage",
        "cross_source_synthesis": "pns_cross_source_synthesis",
        "structural_coverage": "pns_structural_coverage",
        "statistics_and_claims_accuracy": "pns_statistics_and_claims_accuracy",
        "narrative_quality": "pns_narrative_quality"
    },

    "Grant future direction generation": {
        "source_relevance": "grant_source_relevance",
        "novelty": "grant_novelty",
        "feasibility": "grant_feasibility",
        "specificity": "grant_specificity",
        "structure": "grant_structure",
        "metadata_grounding": "grant_metadata_grounding",
        "writing_quality": "grant_writing_quality"
    },

    "Related work generation": {
        "source_grounding": "rw_source_grounding",
        "citation_faithfulness": "rw_citation_faithfulness",
        "reference_coverage": "rw_reference_coverage",
        "synthesis_quality": "rw_synthesis_quality",
        "organization": "rw_organization",
        "specificity": "rw_specificity",
        "writing_quality": "rw_writing_quality"
    }
}
```

```python
common_relationship_metrics = [
    "milestone_achievement_rate",
    "mandatory_milestone_achievement_rate",
    "optional_milestone_achievement_rate",
    "total_milestone_score_normalized",
    "total_dimension_score",
    "total_quality_analysis_score",
    "total_elapsed_sec",
    "coherence",
    "readability",
    "conciseness"
]

multi_agent_columns = [
    "subtask_count",
    "num_agents",
    "planner_elapsed_sec",
    "architect_elapsed_sec",
    "executor_total_elapsed_sec",
    "synthesis_elapsed_sec",
    "judge_elapsed_sec",
    "KPI"
]
```

```python
def build_relationship_dataframe_with_task_prefixes(experiment_dfs):
    records = []

    for (task, configuration, model), df in experiment_dfs.items():
        df = add_analysis_temperature(df)

        if "run_status" in df.columns:
            df = df[
                df["run_status"].eq("successful")
            ].copy()
        else:
            df = df[
                df["total_dimension_score"].notna()
            ].copy()

        df["task"] = task
        df["configuration"] = configuration
        df["model"] = model
        df["row_id"] = df["row_id"].astype(str)

        keep_cols = [
            "task",
            "configuration",
            "model",
            "analysis_temperature",
            "run_id",
            "row_id"
        ]

        available_common_cols = [
            col for col in common_relationship_metrics
            if col in df.columns
        ]

        available_multiagent_cols = [
            col for col in multi_agent_columns
            if col in df.columns
        ]

        rename_map = task_dimension_rename_maps.get(task, {})

        available_task_dimension_cols = [
            col for col in rename_map.keys()
            if col in df.columns
        ]

        selected_cols = (
            keep_cols
            + available_common_cols
            + available_multiagent_cols
            + available_task_dimension_cols
        )

        temp_df = df[selected_cols].copy()

        temp_df = temp_df.rename(
            columns={
                col: rename_map[col]
                for col in available_task_dimension_cols
            }
        )

        numeric_cols = (
            available_common_cols
            + available_multiagent_cols
            + [
                rename_map[col]
                for col in available_task_dimension_cols
            ]
        )

        for col in numeric_cols:
            if col in temp_df.columns:
                temp_df[col] = pd.to_numeric(
                    temp_df[col],
                    errors="coerce"
                )

        if not temp_df.columns.is_unique:
            duplicated_cols = temp_df.columns[
                temp_df.columns.duplicated()
            ].tolist()

            raise ValueError(
                f"Duplicate columns remain after renaming: {duplicated_cols}"
            )

        records.append(
            temp_df.reset_index(drop=True)
        )

    relationship_df = pd.concat(
        records,
        ignore_index=True
    )

    return relationship_df
```

### Build paired Baseline 1 vs Baseline 2 dataframe

```python
import numpy as np
import pandas as pd
from scipy import stats


df_pair_source = analysis_df.copy()


def normalize_baseline_label(x):
    x = str(x).lower()

    if "baseline1" in x or "baseline 1" in x or "single" in x or "direct" in x:
        return "baseline1"

    if "baseline2" in x or "baseline 2" in x or "multi" in x or "decomposed" in x:
        return "baseline2"

    return np.nan


df_pair_source["baseline_key"] = df_pair_source["configuration"].apply(
    normalize_baseline_label
)

df_pair_source = df_pair_source[
    df_pair_source["baseline_key"].isin(["baseline1", "baseline2"])
].copy()

print(df_pair_source["baseline_key"].value_counts())
```

### Define pairing keys and metrics

```python

possible_row_id_columns = [
    "source_row_index",
    "row_index",
    "row_id",
    "grant_id",
    "aid",
    "event"
]

row_id_col = None

for col in possible_row_id_columns:
    if col in df_pair_source.columns:
        row_id_col = col
        break

if row_id_col is None:
    raise ValueError(
        "Could not find a row identifier column. "
        "Expected one of: source_row_index, row_index, row_id, grant_id, aid, event."
    )

pair_keys = [
    "task",
    "model",
    "analysis_temperature",
    "run_id",
    row_id_col
]

print("Using row identifier:", row_id_col)
print("Pairing keys:", pair_keys)


# Define metrics for pairwise comparison

pairwise_metrics = [
    "milestone_achievement_rate",
    "mandatory_milestone_achievement_rate",
    "optional_milestone_achievement_rate",
    "total_milestone_score_normalized",
    "total_dimension_score",
    "total_quality_analysis_score",
    "total_elapsed_sec"
]

pairwise_metrics = [
    col for col in pairwise_metrics
    if col in df_pair_source.columns
]

for col in pairwise_metrics:
    df_pair_source[col] = pd.to_numeric(df_pair_source[col], errors="coerce")

print("Pairwise metrics:")
print(pairwise_metrics)
```

### Create the paired dataframe

```python
# Aggregate duplicates if any, then pivot Baseline 1/Baseline 2

pair_long = (
    df_pair_source
    .groupby(pair_keys + ["baseline_key"], dropna=False)[pairwise_metrics]
    .mean()
    .reset_index()
)

paired_wide = pair_long.pivot_table(
    index=pair_keys,
    columns="baseline_key",
    values=pairwise_metrics,
    aggfunc="mean"
)

paired_wide.columns = [
    f"{metric}_{baseline}"
    for metric, baseline in paired_wide.columns
]

paired_df = paired_wide.reset_index()

required_pair_cols = [
    "total_dimension_score_baseline1",
    "total_dimension_score_baseline2"
]

required_pair_cols = [
    col for col in required_pair_cols
    if col in paired_df.columns
]

if required_pair_cols:
    paired_df = paired_df.dropna(subset=required_pair_cols).copy()

print("Number of paired rows:", len(paired_df))
display(paired_df.head())
```

### Add Baseline 2 − Baseline 1 deltas

```python
# Add deltas: Baseline 2 - Baseline 1

for metric in pairwise_metrics:
    b1_col = f"{metric}_baseline1"
    b2_col = f"{metric}_baseline2"

    if b1_col in paired_df.columns and b2_col in paired_df.columns:
        paired_df[f"{metric}_delta_b2_minus_b1"] = (
            paired_df[b2_col] - paired_df[b1_col]
        )

display(
    paired_df[
        pair_keys
        + [
            col for col in paired_df.columns
            if col.endswith("_delta_b2_minus_b1")
        ]
    ].head()
)
```

### Effects of temperature

```python
# Task-level B2 - B1 comparison by temperature

temperature_task_table = (
    paired_df
    .groupby(["task", "analysis_temperature"], dropna=False)
    .agg(
    milestone_delta=(
        "milestone_achievement_rate_delta_b2_minus_b1",
        "mean"
    ),
    mandatory_milestone_delta=(
        "mandatory_milestone_achievement_rate_delta_b2_minus_b1",
        "mean"
    ),
    optional_milestone_delta=(
        "optional_milestone_achievement_rate_delta_b2_minus_b1",
        "mean"
    ),
    partial_credit_delta=(
        "total_milestone_score_normalized_delta_b2_minus_b1",
        "mean"
    ),
    dimension_delta=(
        "total_dimension_score_delta_b2_minus_b1",
        "mean"
    ),
    quality_delta=(
        "total_quality_analysis_score_delta_b2_minus_b1",
        "mean"
    ),
    runtime_delta_sec=(
        "total_elapsed_sec_delta_b2_minus_b1",
        "mean"
    ),
    n_pairs=(
        "total_dimension_score_delta_b2_minus_b1",
        "count"
    )
)
    .reset_index()
)

temperature_task_table = temperature_task_table.sort_values(
    ["task", "analysis_temperature"]
)

display(temperature_task_table.round(3))
```

```python
# Format table for paper

paper_temperature_task_table = temperature_task_table.copy()

paper_temperature_task_table = paper_temperature_task_table.rename(
    columns={
        "task": "Task",
        "analysis_temperature": "Temp.",
        "milestone_delta": "Overall Milestone Δ",
        "mandatory_milestone_delta": "Mandatory Milestone Δ",
        "optional_milestone_delta": "Optional Milestone Δ",
        "partial_credit_delta": "Partial Credit Δ",
        "dimension_delta": "Dimension Δ",
        "quality_delta": "Quality Δ",
        "runtime_delta_sec": "Runtime Δ (sec)",
        "n_pairs": "N"
    }
)

# Round score columns
for col in [
    "Overall Milestone Δ",
    "Mandatory Milestone Δ",
    "Optional Milestone Δ",
    "Partial Credit Δ",
    "Dimension Δ",
    "Quality Δ"
]:
    paper_temperature_task_table[col] = paper_temperature_task_table[col].map(
        lambda x: f"{x:+.3f}"
    )

# Round runtime column
paper_temperature_task_table["Runtime Δ (sec)"] = paper_temperature_task_table[
    "Runtime Δ (sec)"
].map(lambda x: f"{x:+.1f}")

display(paper_temperature_task_table)
```

### Wilcoxon Signed Test

```python

import numpy as np
import pandas as pd
from scipy.stats import wilcoxon
from statsmodels.stats.multitest import multipletests
from IPython.display import display

delta_col = "total_dimension_score_delta_b2_minus_b1"

# Check required dataframes and variables.

if "model_df" not in globals():
    raise NameError(
        "model_df is not defined. Run the Table 3 preparation cell first."
    )

if "row_id_col" not in globals():
    raise NameError(
        "row_id_col is not defined. Run the pairing-key cell first."
    )

if "model_task_dimension_table" not in globals():
    raise NameError(
        "model_task_dimension_table is not defined. "
        "Run the original Table 3 calculation cell first."
    )

required_cols = [
    "model_label",
    "task_label",
    row_id_col,
    delta_col
]

missing_cols = [
    col for col in required_cols
    if col not in model_df.columns
]

if missing_cols:
    raise KeyError(
        f"Missing required columns in model_df: {missing_cols}\n"
        f"Available columns are:\n{sorted(model_df.columns.tolist())}"
    )

required_table_cols = [
    "model_label",
    "task_label",
    delta_col
]

missing_table_cols = [
    col for col in required_table_cols
    if col not in model_task_dimension_table.columns
]

if missing_table_cols:
    raise KeyError(
        "The original Table 3 dataframe is missing these columns: "
        f"{missing_table_cols}\n"
        "Run the original Table 3 calculation cell before this cell."
    )

print(f"Using source-instance column: {row_id_col}")

model_task_test_df = model_df[required_cols].copy()

model_task_test_df[delta_col] = pd.to_numeric(
    model_task_test_df[delta_col],
    errors="coerce"
)

model_task_test_df = model_task_test_df.dropna(
    subset=[
        "model_label",
        "task_label",
        row_id_col,
        delta_col
    ]
)


source_level_delta = (
    model_task_test_df
    .groupby(
        ["model_label", "task_label", row_id_col],
        dropna=False
    )[delta_col]
    .mean()
    .reset_index(name="source_mean_delta_tqs")
)

print("Source-level paired differences used by the Wilcoxon tests:")
display(source_level_delta.head())

# Run one Wilcoxon signed-rank test per model-task pair.


results = []

for (model_name, task_name), group in source_level_delta.groupby(
    ["model_label", "task_label"],
    dropna=False
):
    deltas = (
        group["source_mean_delta_tqs"]
        .replace([np.inf, -np.inf], np.nan)
        .dropna()
        .to_numpy()
    )

    n_sources = len(deltas)

    source_level_mean_delta = (
        deltas.mean()
        if n_sources > 0
        else np.nan
    )

    source_level_median_delta = (
        np.median(deltas)
        if n_sources > 0
        else np.nan
    )

    if n_sources == 0:
        statistic = np.nan
        p_value = np.nan
        test_status = "No observations"

    elif np.allclose(deltas, 0):
        statistic = 0.0
        p_value = 1.0
        test_status = "All differences zero"

    else:
        try:
            test = wilcoxon(
                deltas,
                alternative="two-sided",
                zero_method="wilcox",
                correction=False,
                method="auto"
            )

            statistic = test.statistic
            p_value = test.pvalue
            test_status = "Successful"

        except ValueError as error:
            statistic = np.nan
            p_value = np.nan
            test_status = str(error)

    results.append({
        "model_label": model_name,
        "task_label": task_name,
        "n_sources": n_sources,
        "source_level_mean_delta_tqs": source_level_mean_delta,
        "source_level_median_delta_tqs": source_level_median_delta,
        "wilcoxon_statistic": statistic,
        "p_value_raw": p_value,
        "test_status": test_status
    })

model_task_wilcoxon = pd.DataFrame(results)

# Apply Holm correction across all model-task tests.

valid_mask = model_task_wilcoxon["p_value_raw"].notna()

model_task_wilcoxon["p_value_holm"] = np.nan
model_task_wilcoxon["significant_holm_0_05"] = False

if valid_mask.any():
    reject, adjusted_p, _, _ = multipletests(
        model_task_wilcoxon.loc[
            valid_mask,
            "p_value_raw"
        ],
        alpha=0.05,
        method="holm"
    )

    model_task_wilcoxon.loc[
        valid_mask,
        "p_value_holm"
    ] = adjusted_p

    model_task_wilcoxon.loc[
        valid_mask,
        "significant_holm_0_05"
    ] = reject



original_table_means = (
    model_task_dimension_table[
        [
            "model_label",
            "task_label",
            delta_col
        ]
    ]
    .copy()
    .rename(
        columns={
            delta_col: "reported_mean_delta_tqs"
        }
    )
)

model_task_wilcoxon = model_task_wilcoxon.merge(
    original_table_means,
    on=["model_label", "task_label"],
    how="left",
    validate="one_to_one"
)

if model_task_wilcoxon["reported_mean_delta_tqs"].isna().any():
    missing_original_means = model_task_wilcoxon.loc[
        model_task_wilcoxon["reported_mean_delta_tqs"].isna(),
        ["model_label", "task_label"]
    ]

    raise ValueError(
        "Some Wilcoxon results could not be matched to the "
        "original Table 3 means:\n"
        f"{missing_original_means}"
    )

# Add Holm-corrected significance markers.

def significance_marker(p_value):
    if pd.isna(p_value):
        return ""

    if p_value < 0.001:
        return "***"

    if p_value < 0.01:
        return "**"

    if p_value < 0.05:
        return "*"

    return ""


model_task_wilcoxon["significance"] = (
    model_task_wilcoxon["p_value_holm"]
    .apply(significance_marker)
)

model_task_wilcoxon["table_value"] = (
    model_task_wilcoxon.apply(
        lambda row: (
            f"{row['reported_mean_delta_tqs']:+.4f}"
            f"{row['significance']}"
        ),
        axis=1
    )
)

print("Full Wilcoxon results:")
display(
    model_task_wilcoxon
    .sort_values(["task_label", "model_label"])
    .round(4)
)

# Create the paper-ready model x task table.

model_task_significance_table = (
    model_task_wilcoxon
    .pivot(
        index="model_label",
        columns="task_label",
        values="table_value"
    )
)

preferred_model_order = [
    "Gemma3",
    "Qwen3:14B",
    "DeepSeek-R1:32B",
    "GPT-5-mini"
]

preferred_task_order = [
    "Public Narrative",
    "Project Ideation",
    "Related Work"
]

model_task_significance_table = (
    model_task_significance_table.reindex(
        index=[
            model
            for model in preferred_model_order
            if model in model_task_significance_table.index
        ],
        columns=[
            task
            for task in preferred_task_order
            if task in model_task_significance_table.columns
        ]
    )
)

print(
    "Table 3 with the original means and "
    "Holm-corrected Wilcoxon significance markers:"
)

display(model_task_significance_table)


model_task_significance_table.to_csv(
    "table3_model_task_tqs_wilcoxon_formatted.csv"
)

model_task_wilcoxon.to_csv(
    "table3_model_task_tqs_wilcoxon_full_results.csv",
    index=False
)

print(
    "The displayed means come from the original Table 3 "
    "aggregation. Wilcoxon tests use source-level paired "
    "differences. Significance markers use Holm-adjusted "
    "p-values: * p<0.05, ** p<0.01, *** p<0.001."
)
```

```python

def plot_executor_agent_relationship(
    executor_df,
    save=True,
    save_name="executor_agent_metric_relationship"
):
    plot_metrics = [
        "milestone_achievement_rate",
        "total_dimension_score",
        "total_quality_analysis_score"
    ]

    plot_metrics = [
        metric for metric in plot_metrics
        if metric in executor_df.columns
    ]

    metric_labels = {
        "milestone_achievement_rate": "MAR",
        "total_dimension_score": "TQS",
        "total_quality_analysis_score": "CQS"
    }

    fig, axes = plt.subplots(
        1,
        len(plot_metrics),
        figsize=(4.0 * len(plot_metrics), 3.8),
        sharex=True
    )

    if len(plot_metrics) == 1:
        axes = [axes]

    rng = np.random.default_rng(42)

    for ax, metric in zip(axes, plot_metrics):
        valid = executor_df[["num_agents", metric]].dropna().copy()

        # Jitter only for readability
        x_jitter = valid["num_agents"] + rng.normal(
            0,
            0.035,
            size=len(valid)
        )

        ax.scatter(
            x_jitter,
            valid[metric],
            alpha=0.25,
            s=18
        )

        # Mean line by number of agents
        mean_by_agents = (
            valid
            .groupby("num_agents")[metric]
            .mean()
            .reset_index()
            .sort_values("num_agents")
        )

        ax.plot(
            mean_by_agents["num_agents"],
            mean_by_agents[metric],
            marker="o",
            linewidth=2.2
        )

        ax.set_title(metric_labels.get(metric, metric))
        ax.set_xlabel("Number of executor agents")
        ax.grid(axis="y", alpha=0.3)

        if metric != "total_elapsed_sec":
            ax.set_ylim(0, 1.05)

    axes[0].set_ylabel("Metric value")


    plt.tight_layout()

    if save:
        os.makedirs("executor_agent_analysis_figures", exist_ok=True)
        plt.savefig(
            f"executor_agent_analysis_figures/{save_name}.png",
            dpi=300,
            bbox_inches="tight"
        )
        plt.savefig(
            f"executor_agent_analysis_figures/{save_name}.pdf",
            bbox_inches="tight"
        )

    plt.show()


plot_executor_agent_relationship(executor_df)
```

### Three Panel Dumble Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


if "analysis_df" not in globals():
    raise NameError(
        "analysis_df is not defined. Run the cleaning cell first, especially:\n"
        "analysis_df = relationship_df_clean.copy()"
    )

plot_source_df = analysis_df.copy()


# Normalize workflow labels

def normalize_workflow_label(x):
    x = str(x).lower()

    if "baseline1" in x or "baseline 1" in x or "single" in x or "direct" in x:
        return "Direct"

    if "baseline2" in x or "baseline 2" in x or "multi" in x or "decomposed" in x:
        return "Decomposed"

    return np.nan


plot_source_df["workflow"] = plot_source_df["configuration"].apply(
    normalize_workflow_label
)

plot_source_df = plot_source_df[
    plot_source_df["workflow"].isin(["Direct", "Decomposed"])
].copy()

print("Workflow counts:")
print(plot_source_df["workflow"].value_counts())
```

```python

task_metric_map = {
    "Public-event narrative synthesis": {
        "metrics": [
            "pns_core_event_fidelity",
            "pns_required_entity_coverage",
            "pns_key_information_coverage",
            "pns_cross_source_synthesis",
            "pns_structural_coverage",
            "pns_statistics_and_claims_accuracy",
            "pns_narrative_quality",
            "total_quality_analysis_score"
        ],
        "labels": [
            "Core event fidelity",
            "Required entity coverage",
            "Key information coverage",
            "Cross-source synthesis",
            "Structural coverage",
            "Statistics/claims accuracy",
            "Narrative quality",
            "Writing quality score"
        ]
    },

    "Grant future direction generation": {
        "metrics": [
            "grant_source_relevance",
            "grant_novelty",
            "grant_feasibility",
            "grant_specificity",
            "grant_structure",
            "grant_metadata_grounding",
            "grant_writing_quality",
            "total_quality_analysis_score"
        ],
        "labels": [
            "Source relevance",
            "Novelty",
            "Feasibility",
            "Specificity",
            "Structure",
            "Metadata grounding",
            "Writing quality",
            "Writing quality score"
        ]
    },

    "Related work generation": {
        "metrics": [
            "rw_source_grounding",
            "rw_citation_faithfulness",
            "rw_reference_coverage",
            "rw_synthesis_quality",
            "rw_organization",
            "rw_specificity",
            "rw_writing_quality",
            "total_quality_analysis_score"
        ],
        "labels": [
            "Source grounding",
            "Citation faithfulness",
            "Reference coverage",
            "Synthesis quality",
            "Organization",
            "Specificity",
            "Writing quality",
            "Writing quality score"
        ]
    }
}
```

```python
# Build Direct vs. Decomposed summary table

summary_rows = []

for task_name, info in task_metric_map.items():
    task_df = plot_source_df[
        plot_source_df["task"] == task_name
    ].copy()

    if task_df.empty:
        print(f"Warning: no rows found for task: {task_name}")
        continue

    metric_cols = [
        col for col in info["metrics"]
        if col in task_df.columns
    ]

    metric_labels = [
        label
        for col, label in zip(info["metrics"], info["labels"])
        if col in metric_cols
    ]

    if len(metric_cols) == 0:
        print(f"Warning: no metric columns found for task: {task_name}")
        continue

    for col in metric_cols:
        task_df[col] = pd.to_numeric(task_df[col], errors="coerce")

    group_mean = (
        task_df
        .groupby(
            [
                "workflow",
                "model",
                "analysis_temperature",
                "run_id"
            ],
            dropna=False
        )[metric_cols]
        .mean()
        .reset_index()
    )

    # Then average across models, temperatures, and runs
    workflow_mean = (
        group_mean
        .groupby("workflow", dropna=False)[metric_cols]
        .mean()
        .reset_index()
    )

    for metric_col, metric_label in zip(metric_cols, metric_labels):
        direct_value = workflow_mean.loc[
            workflow_mean["workflow"] == "Direct",
            metric_col
        ]

        decomposed_value = workflow_mean.loc[
            workflow_mean["workflow"] == "Decomposed",
            metric_col
        ]

        if len(direct_value) == 0 or len(decomposed_value) == 0:
            continue

        direct_value = direct_value.iloc[0]
        decomposed_value = decomposed_value.iloc[0]

        summary_rows.append({
            "task": task_name,
            "metric": metric_col,
            "metric_label": metric_label,
            "Direct": direct_value,
            "Decomposed": decomposed_value,
            "Delta": decomposed_value - direct_value
        })

dimension_quality_dumbbell_df = pd.DataFrame(summary_rows)

display(dimension_quality_dumbbell_df.round(3))

dimension_quality_dumbbell_df.to_csv(
    "dimension_quality_dumbbell_summary.csv",
    index=False
)
```

```python
# Plot three-panel dumbbell figure

def plot_three_panel_dumbbell(
    summary_df,
    task_order,
    save_path="three_panel_dimension_quality_dumbbell_clean.png"
):
    fig, axes = plt.subplots(
        1,
        3,
        figsize=(18, 7),
        sharex=True
    )

    for ax, task_name in zip(axes, task_order):
        task_plot = summary_df[
            summary_df["task"] == task_name
        ].copy()

        if task_plot.empty:
            ax.set_title(f"{task_name}\nNo data")
            continue

        # Reverse order so first metric appears at top
        task_plot = task_plot.iloc[::-1].reset_index(drop=True)

        y_positions = np.arange(len(task_plot))

        for i, row in task_plot.iterrows():
            direct_score = row["Direct"]
            decomposed_score = row["Decomposed"]
            delta = row["Delta"]

            # Connecting line: one consistent style
            ax.plot(
                [direct_score, decomposed_score],
                [i, i],
                color="0.45",
                linewidth=2.5,
                alpha=0.9,
                zorder=1
            )

            # Direct: hollow circle
            ax.scatter(
                direct_score,
                i,
                s=80,
                facecolors="white",
                edgecolors="black",
                linewidths=1.7,
                marker="o",
                label="Direct" if i == 0 else "",
                zorder=3
            )

            # Decomposed: filled circle
            ax.scatter(
                decomposed_score,
                i,
                s=80,
                facecolors="black",
                edgecolors="black",
                linewidths=1.7,
                marker="o",
                label="Decomposed" if i == 0 else "",
                zorder=4
            )

            # Delta label
            x_text = max(direct_score, decomposed_score) + 0.008

            ax.text(
                x_text,
                i,
                f"{delta:+.3f}",
                va="center",
                fontsize=8
            )

        ax.set_yticks(y_positions)
        ax.set_yticklabels(task_plot["metric_label"], fontsize=9)

        ax.set_xlim(0.75, 1.03)
        ax.set_xticks([0.75, 0.80, 0.85, 0.90, 0.95, 1.00])

        ax.set_xlabel("Mean score")
        
        display_task_name = {
        "Grant future direction generation": "Project-grounded Research Ideation",
        "Public-event narrative synthesis": "Public-event Narrative Synthesis",
        "Related work generation": "Related Work Generation"
    }.get(task_name, task_name)

        ax.set_title(display_task_name, fontsize=11)

        ax.grid(axis="x", alpha=0.25)

        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)

    handles, labels = axes[0].get_legend_handles_labels()

    fig.legend(
        handles,
        labels,
        loc="lower center",
        ncol=2,
        frameon=False,
        fontsize=11
    )

    fig.suptitle(
        "Task-specific dimension and quality-analysis scores",
        fontsize=14,
        y=1.02
    )

    plt.tight_layout(rect=[0, 0.07, 1, 1])

    plt.savefig(
        save_path,
        dpi=300,
        bbox_inches="tight"
    )

    plt.show()

task_order = [
    "Public-event narrative synthesis",
    "Grant future direction generation",
    "Related work generation"
]

plot_three_panel_dumbbell(
    summary_df=dimension_quality_dumbbell_df,
    task_order=task_order,
    save_path="three_panel_dimension_quality_dumbbell.png"
)
```

```python


summary_rows = []

for task_name, info in task_metric_map.items():
    task_df = plot_source_df[
        plot_source_df["task"] == task_name
    ].copy()

    if task_df.empty:
        print(f"Warning: no rows found for task: {task_name}")
        continue

    available_items = [
        (metric, label, group)
        for metric, label, group in zip(
            info["metrics"],
            info["labels"],
            info["groups"]
        )
        if metric in task_df.columns
    ]

    if not available_items:
        print(f"Warning: no metric columns found for task: {task_name}")
        continue

    metric_cols = [item[0] for item in available_items]

    for col in metric_cols:
        task_df[col] = pd.to_numeric(task_df[col], errors="coerce")

    # First average within workflow-model-temperature-run.
    # This prevents one model/run/temperature group from dominating.
    group_mean = (
        task_df
        .groupby(
            [
                "workflow",
                "model",
                "analysis_temperature",
                "run_id"
            ],
            dropna=False
        )[metric_cols]
        .mean()
        .reset_index()
    )

    # Then average across models, temperatures, and runs.
    workflow_mean = (
        group_mean
        .groupby("workflow", dropna=False)[metric_cols]
        .mean()
        .reset_index()
    )

    for metric_col, metric_label, metric_group in available_items:
        direct_value = workflow_mean.loc[
            workflow_mean["workflow"] == "Direct",
            metric_col
        ]

        decomposed_value = workflow_mean.loc[
            workflow_mean["workflow"] == "Decomposed",
            metric_col
        ]

        if direct_value.empty or decomposed_value.empty:
            continue

        direct_value = direct_value.iloc[0]
        decomposed_value = decomposed_value.iloc[0]

        summary_rows.append({
            "task": task_name,
            "metric_group": metric_group,
            "metric": metric_col,
            "metric_label": metric_label,
            "Direct": direct_value,
            "Decomposed": decomposed_value,
            "Delta": decomposed_value - direct_value
        })

granular_tqs_cqs_dumbbell_df = pd.DataFrame(summary_rows)

display(granular_tqs_cqs_dumbbell_df.round(3))

granular_tqs_cqs_dumbbell_df.to_csv(
    "granular_tqs_cqs_dumbbell_summary.csv",
    index=False
)
```

### Time–quality tradeoff metrics

```python


QUALITY_METRIC = "total_dimension_score"
TIME_METRIC = "total_elapsed_sec"

quality_b1 = f"{QUALITY_METRIC}_baseline1"
quality_b2 = f"{QUALITY_METRIC}_baseline2"
time_b1 = f"{TIME_METRIC}_baseline1"
time_b2 = f"{TIME_METRIC}_baseline2"

required_cols = [quality_b1, quality_b2, time_b1, time_b2]

missing_cols = [
    col for col in required_cols
    if col not in paired_df.columns
]

if missing_cols:
    raise ValueError(f"Missing required tradeoff columns: {missing_cols}")


# Avoid division by zero
EPS = 1e-9

paired_df["quality_delta_b2_minus_b1"] = (
    paired_df[quality_b2] - paired_df[quality_b1]
)

paired_df["time_delta_b2_minus_b1"] = (
    paired_df[time_b2] - paired_df[time_b1]
)

paired_df["time_delta_minutes_b2_minus_b1"] = (
    paired_df["time_delta_b2_minus_b1"] / 60
)

paired_df["quality_per_second_baseline1"] = (
    paired_df[quality_b1] / (paired_df[time_b1] + EPS)
)

paired_df["quality_per_second_baseline2"] = (
    paired_df[quality_b2] / (paired_df[time_b2] + EPS)
)

paired_df["quality_per_second_delta_b2_minus_b1"] = (
    paired_df["quality_per_second_baseline2"]
    - paired_df["quality_per_second_baseline1"]
)

paired_df["relative_quality_change_pct"] = (
    paired_df["quality_delta_b2_minus_b1"]
    / (paired_df[quality_b1] + EPS)
) * 100

paired_df["relative_time_change_pct"] = (
    paired_df["time_delta_b2_minus_b1"]
    / (paired_df[time_b1] + EPS)
) * 100

# Quality gain per extra minute.
# If Baseline 2 is faster, this value is still computed,
# but interpretation is strongest when time_delta > 0.
paired_df["quality_delta_per_extra_minute"] = (
    paired_df["quality_delta_b2_minus_b1"]
    / (paired_df["time_delta_minutes_b2_minus_b1"] + EPS)
)



# Interpretation:
# A penalty of 0.01 means each extra minute costs 0.01 quality points.
# Since quality is 0-1, this is easy to explain in the paper.
# You can test 0.005, 0.01, and 0.02 as sensitivity checks.

TIME_PENALTY_PER_MINUTE = 0.01

paired_df["time_penalized_quality_baseline1"] = (
    paired_df[quality_b1]
    - TIME_PENALTY_PER_MINUTE * (paired_df[time_b1] / 60)
)

paired_df["time_penalized_quality_baseline2"] = (
    paired_df[quality_b2]
    - TIME_PENALTY_PER_MINUTE * (paired_df[time_b2] / 60)
)

paired_df["time_penalized_quality_delta_b2_minus_b1"] = (
    paired_df["time_penalized_quality_baseline2"]
    - paired_df["time_penalized_quality_baseline1"]
)
```

### Summarize tradeoff metrics

```python
quality_time_tradeoff = (
    paired_df
    .groupby("task", dropna=False)
    .agg(
        mean_quality_delta=("quality_delta_b2_minus_b1", "mean"),
        mean_runtime_delta_sec=("time_delta_b2_minus_b1", "mean"),
        n_pairs=("quality_delta_b2_minus_b1", "count")
    )
    .reset_index()
)

quality_time_tradeoff["quality_gain_per_extra_minute"] = (
    quality_time_tradeoff["mean_quality_delta"]
    / (quality_time_tradeoff["mean_runtime_delta_sec"] / 60)
)

display(quality_time_tradeoff.round(4))
```
